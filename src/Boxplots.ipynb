# BoxPlots

```python
import glob
inputs = ['../results/csv_based/SlidingWindowParameters/',
           '../results/csv_without_interms/SlidingWindowParameters/']

ind = 1

root = inputs[ind]

csvs = glob.glob(root + '/*.csv')
csvs.sort()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

dfs = []

for c in csvs:
    print(c)
    dataframe = pd.read_csv(c)
    dfs.append(dataframe)
```

```python
print(len(dfs))
#print(dfs[5])
combined_frame = pd.concat(dfs)
new_df = dfs[5].sort_values(by=["skill", "method"])
print(new_df)
#cnn_values = new_df.loc[new_df['method'] == 'CNN']
#print(cnn_values)
```

## ResNet Boxplot

```python
#Specific Example of 180x60
resnet_values = new_df.loc[new_df['method'] == 'ResNet']
resnet_knot = resnet_values.loc[resnet_values['skill']=='Knot']
resnet_needle = resnet_values.loc[resnet_values['skill']=='Needle']
resnet_suture = resnet_values.loc[resnet_values['skill']=='Suture']

# prepare data
data_to_plot = [resnet_needle['Acc_Full'], resnet_needle['Acc_Fea'],
                resnet_knot['Acc_Full'], resnet_knot['Acc_Fea'],
                resnet_suture['Acc_Full'], resnet_suture['Acc_Fea']]

fig = plt.figure(figsize=(10, 6))
plt.ylim(0.0,1.0)
fig.suptitle('ResNet accuracies by skills - 180x60 (without interms)')
box = plt.boxplot(data_to_plot,
                  labels=['Needle_Model','Needle_SVM', 'Knot_Model', 'Knot_SVM','Suture_Model', 'Suture_SVM'])
```

```python
#Combined
resnet_values = combined_frame.loc[combined_frame['method'] == 'ResNet']
resnet_knot = resnet_values.loc[resnet_values['skill']=='Knot']
resnet_needle = resnet_values.loc[resnet_values['skill']=='Needle']
resnet_suture = resnet_values.loc[resnet_values['skill']=='Suture']

# prepare data
data_to_plot = [resnet_needle['Acc_Full'], resnet_needle['Acc_Fea'],
                resnet_knot['Acc_Full'], resnet_knot['Acc_Fea'],
                resnet_suture['Acc_Full'], resnet_suture['Acc_Fea']]

fig = plt.figure(figsize=(10, 6))
plt.ylim(0.0,1.0)
fig.suptitle('ResNet accuracies by skills - all data (without interms)')
box = plt.boxplot(data_to_plot,
                  labels=['Needle_Model','Needle_SVM', 'Knot_Model', 'Knot_SVM','Suture_Model', 'Suture_SVM'])
```
